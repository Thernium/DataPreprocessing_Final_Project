### **1. Which department has the highest average performance?**

```python
department_performance = (
    Employees.groupby("Department")["Performance Score"]
    .mean()
    .sort_values(ascending=False)
)

print(department_performance)
```

#### **Answer: 1** HR

### **2. Which department has the highest salary?**

```python
department_salary = (
    Employees.groupby("Department")["Salary"]
    .mean()
    .sort_values(ascending=False)
)

print(department_salary)
```

#### **Answer: 2** Engineering

### **3. Which education level performs best?**

```python
education_performance = (
    Employees.groupby("Education")["Performance Score"]
    .mean()
    .sort_values(ascending=False)
)

print(education_performance)
```

#### **Answer 3:** PhD

### **4. Which age group performs best?**

```python
age_performance = (
    Employees.groupby("Age")["Performance Score"]
    .mean()
    .sort_values(ascending=False)
)

print(age_performance)
```

#### **Answer: 4** 50-60

### **5. Is attendance related to performance?**

```python
Employees[["Attendance Rate", "Performance Score"]].corr()
```

#### **Answer 5:** Yes

### **6. Do more projects always improve performance?**

```python
Employees[["Projects Completed", "Performance Score"]].corr()
```

#### **Answer 6:** Yes

### **7. Which variables have the strongest positive correlation?**

```python
corr = Employees.corr(numeric_only=True)

strongest_positive = (
    corr.where(~np.eye(corr.shape[0], dtype=bool))
        .stack()
        .sort_values(ascending=False)
)

print(strongest_positive.head())
```

#### **Answer:** 
Experience - ProjectsCompleted <br>
Experience - Performance Score <br>
...


### **8. Which variables have the strongest negative correlation?**

```python
strongest_negative = (
    corr.where(~np.eye(corr.shape[0], dtype=bool))
        .stack()
        .sort_values()
)

print(strongest_negative.head())
```

#### **Answer:** 
TrainingHours - Experience <br>
TrainingHours - Age <br>
...
