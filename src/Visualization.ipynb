# **Phase 7 - Visualization (50-57)**
Histogram, Bar Chart, Pie Chart, Scatter Plot, Box Plot, Count Plot, Violin Plot, Correlation
Heatmap.

```python
import pandas as pd

Employees = pd.read_excel("../Employees.xlsx")
```

## Importation

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
print(Employees.columns)
```

```python
# Pie Chart, Scatter Plot, Box Plot, Count Plot, Violin Plot, Correlation
# Heatmap.
```

## Histogram

```python
sns.histplot(
    Employees["Performance Score"],
    bins=42
)

plt.title("Performance Score Distribution")
plt.xlabel("Performance Score")
plt.ylabel("Number of Employees")

plt.show()
```

```python
sns.histplot(
    Employees["Age"],
    bins=40,
)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Employees")

plt.show()
```

## Bar Chart

```python
#Data
avg_salary = Employees.groupby("Department")["Salary"].mean()

sns.barplot(
    x=avg_salary.index,
    y=avg_salary.values
)

plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")

plt.show()
```

## Pie Chart

```python
#Data
department_counts = Employees["Education"].value_counts()

plt.pie(
    department_counts,
    labels=department_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Employee Distribution by Education")


plt.show()
```

## Scatter

```python
sns.scatterplot(
    x = Employees["Experience"],
    y = Employees["Performance Score"]
)

plt.show()
```

```python
sns.scatterplot(
    x = Employees["Experience"],
    y = Employees["Salary"]
)

plt.show()
```

```python
sns.scatterplot(
    x = Employees["Age"],
    y = Employees["Salary"]
)

plt.show()
```

## Box Plot

```python
sns.boxplot(
    data=Employees,
    x="Department",
    y="Salary"
)

plt.title("Salary Distribution by Department")

plt.xticks(rotation=20)

plt.show()
```

Count Plot

```python
sns.countplot(
    data=Employees,
    x="Education"
)

plt.title("Employees by Education Level")

plt.xticks(rotation=15)

plt.show()
```

## Violin Plot

```python
sns.violinplot(
    data=Employees,
    x="Department",
    y="Performance Score"
)

plt.title("Performance Score Distribution by Department")

plt.show()
```

## Correlation Heatmap

```python
#Data
corr = Employees.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.show()
```
