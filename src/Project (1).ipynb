```python
#Employee Experience
employee_experience = np.array([np.random.randint(0, age - 18) for age in employee_age])
```

5. Generate Salary using Experience and Department.

```python
department_base_salary = {
    "Engineering": (60000, 90000),
    "Sales":       (40000, 65000),
    "Marketing":   (45000, 70000),
    "HR":          (40000, 60000),
    "Finance":     (55000, 80000),
}
```

```python
#Employee Salary
employee_salary = np.array([
    int(np.random.randint(*department_base_salary[dept]) + (exp * 1000))
    for dept, exp in zip(employee_department, employee_experience)
])
# print(f"{len(employee_salary)} \n{employee_salary}")
```

6. Generate Training Hours.

```python
#Employee Training Hours
employee_training_hours = np.array([
    np.random.randint(10, 80) if exp < 5 else np.random.randint(5, 40)
    for exp in employee_experience
])
```

7. Generate Projects Completed.

```python
#Employee Projects Completed
employee_projects_completed = np.array([
    np.random.randint(exp*6, exp*12) if exp > 0 else 0
    for exp in employee_experience
])
```

8. Generate Attendance Rate.

```python
#Employee Attendance Rate
employee_attendance_rate = np.round(
    np.random.uniform(0.75, 1.0, size=500), 2
)
```

9. Generate Department and **Education**.<br>
----------------------------- ^^^^^^^

```python
#Employee Education
employee_education = np.random.choice(
    ["High School", "Bachelor's", "Master's", "PhD"],
    size=500,
    p=[0.01, 0.60, 0.32, 0.07]
)
```

**10. Compute Performance Score using multiple variables.**

```python
#Employee Performance Score

# normalize each component to 0-1 range first
norm_experience = employee_experience / employee_experience.max()
norm_training   = employee_training_hours / employee_training_hours.max()
norm_attendance = employee_attendance_rate

# education score
education_score_map = {
    "High School": 0.1,
    "Bachelor's":  0.6,
    "Master's":    0.8,
    "PhD":         1.0,
}
norm_education = np.array([education_score_map[e] for e in employee_education])

# weighted formula
employee_performance_score = np.round(
    (norm_experience  * 0.30 +
     norm_attendance  * 0.30 +
     norm_training    * 0.20 +
     norm_education   * 0.20) * 100,
    2
)

print(f"Max: %{max(employee_performance_score)}")
print(f"Min: %{min(employee_performance_score)}")
```

## **Inserting Problems**

```python
age_indexes_to_delete = np.random.randint(1, 500, size=4) # Detecting 4 random indexes to delete from the "Age" column
age_2_indexes_to_delete = np.random.randint(1, 500, size=3) # Detecting 3 random indexes to delete from the "Age" column
dprt_indexes_to_delete = np.random.randint(1, 500, size=3) # Detecting 3 random indexes to delete from the "Department" column
exp_indexes_to_delete = np.random.randint(1, 500, size=4) # Detecting 4 random indexes to delete from the "Experience" column
slry_indexes_to_delete = np.random.randint(1, 500, size=2) # Detecting 2 random indexes to delete from the "Salary" column
slry_2_indexes_to_delete = np.random.randint(1, 500, size=1) # Detecting 1 random indexes to delete from the "Salary" column
performance_indexes_to_delete = np.random.randint(1, 500, size=2) # Detecting 2 random indexes to delete from the "Performance Score" column
performance_2_indexes_to_delete = np.random.randint(1, 500, size=1) # Detecting 1 random indexes to delete from the "Performance Score" column

#Deleting
Employees.iloc[age_indexes_to_delete, Employees.columns.get_loc("Age")] = np.nan
Employees.iloc[age_2_indexes_to_delete, Employees.columns.get_loc("Age")] = [10, 15, -2]

Employees.iloc[dprt_indexes_to_delete, Employees.columns.get_loc("Department")] = np.nan
Employees.iloc[exp_indexes_to_delete, Employees.columns.get_loc("Experience")] = np.nan

Employees.iloc[slry_indexes_to_delete, Employees.columns.get_loc("Salary")] = np.nan
Employees.iloc[slry_2_indexes_to_delete, Employees.columns.get_loc("Salary")] *= -1

Employees.iloc[performance_indexes_to_delete, Employees.columns.get_loc("Performance Score")] = np.nan
Employees.iloc[performance_2_indexes_to_delete, Employees.columns.get_loc("Performance Score")] += 200

# print(Employees.iloc[age_2_indexes_to_delete, Employees.columns.get_loc("Age")] )


#Inserting Duplicates
duplicates_indexes = np.random.randint(0, 500, size=3) #Random Indexes to Duplicate

duplicated_rows = Employees.iloc[duplicates_indexes]
Employees = pd.concat([Employees, duplicated_rows], ignore_index=False)
# print(Employees)
```

```python
#Inserting Inconsistent Categories
Employees['department_dirty'] = Employees['Department'].apply(
    lambda x: np.random.choice([x.lower(), x.upper(), x.title(), x]) if pd.notna(x) else x
)
print(Employees['department_dirty'])

print(Employees["Department"].isnull().any())

```

## **Fixing The Problems We Created**

```python
#Detecting Missing Values
Employees_missing_values = Employees.isnull()
# print(Employees_missing_values.sum())

#Age Missing & Invalid Values
    #Set Missing Values -----> nan
Employees.loc[~Employees["Age"].between(18, 65), "Age"] = np.nan
Age_missing_values = Employees_missing_values["Age"]
Employees.loc[Age_missing_values, "Age"] = Employees["Age"].mean()

print(f"Missing/Invalid Age Values? =>{Employees.isna()["Age"].any()}")

#Department Missing Values
title_to_department = {
    title: department
    for department, titles in departments_titles.items()
    for title in titles
}

Employees["Department"] = Employees["Department"].fillna(
    Employees["Job Title"].map(title_to_department)
)

#Experience Missing Values
Employees["Age Group"] = pd.cut(
    Employees["Age"],
    bins=[18, 25, 35, 45, 55, 65],
)

Employees["Experience"] = Employees["Experience"].fillna(
    Employees.groupby("Age Group")["Experience"].transform("median")
)

Employees.drop(columns="Age Group", inplace=True)

# print(f"Is there any Missing Values in 'Experience': {Employees.isna()["Experience"].any()}")

#Salary Missing & Invalid Values
    #Set Missing Values -----> nan
Employees[Employees["Salary"] <= 0] = np.nan
Employees["Salary"] = Employees["Salary"].fillna(
    Employees.groupby("Job Title")["Salary"].transform("median")
)

print(f"Is there any Missing Values in 'Salary': {Employees.isna()["Salary"].any()}")
```

## Performance Missing & Invalid Values

```python
#Set Invalids -----> nan
Employees[(Employees["Performance Score"] > 100) | (Employees["Performance Score"] < 0)] = np.nan

#Filling Performance Missing Values 
mask = Employees["Performance Score"].isna()

education_score_map = {
    "High School": 0.1,
    "Bachelor's": 0.6,
    "Master's": 0.8,
    "PhD": 1.0,
}

Employees.loc[mask, "Performance Score"] = np.round(
    (
        (Employees.loc[mask, "Experience"] / Employees["Experience"].max()) * 0.30 +
        Employees.loc[mask, "Attendance Rate"] * 0.30 +
        (Employees.loc[mask, "Training Hours"] / Employees["Training Hours"].max()) * 0.20 +
        Employees.loc[mask, "Education"].map(education_score_map) * 0.20
    ) * 100,
    2
)
```

# **Phase 5 - Feature Engineering (34-41)**
34. AgeGroup.
35. SalaryLevel.
36. ExperienceLevel.
37. TrainingCategory.
38. AttendanceCategory.
39. HighPerformer flag.
40. BonusEligibility.
41. Rearrange columns.

```python
#34.
Employees["AgeGroup"] = pd.cut(
    Employees["Age"],
    bins=[18, 30, 45, 60, 65],
    labels=["Young", "Mid-Career", "Senior", "Near Retirement"],
    include_lowest=True
)

#35.
Employees["SalaryLevel"] = pd.qcut(
    Employees["Salary"],
    q=3,
    labels=["Low", "Medium", "High"]
)

#36.
Employees["ExperienceLevel"] = pd.cut(
    Employees["Experience"],
    bins=[-1, 2, 5, 10, np.inf],
    labels=[
        "Entry",
        "Junior",
        "Mid-Level",
        "Senior"
    ]
)

#37.
Employees["TrainingCategory"] = pd.qcut(
    Employees["Training Hours"],
    q=3,
    labels=["Low", "Medium", "High"]
)

#38.
Employees["AttendanceCategory"] = pd.cut(
    Employees["Attendance Rate"],
    bins=[0, 0.5, 0.80, 0.95, 1.0],
    labels=["Poor", "Not-Good", "Good", "Excellent"],
    include_lowest=True
)

#39.
# print(max(Employees["Performance Score"]))
# print(f"HighPerformer employees: %{round((len(Employees[Employees["Performance Score"] >= 60])) / len(Employees) * 100, 2)}")
Employees["HighPerformer"] = (
    Employees["Performance Score"] >= 60
)

#40.
Employees["BonusEligibility"] = (
    (Employees["Performance Score"] >= 60) &
    (Employees["Attendance Rate"] >= 0.95) &
    (Employees["Experience"] >= 3)
)
# print(f"BonusEligibility employees: {Employees["BonusEligibility"].sum()}")

#41.
Employees = Employees[
    [
        "First Name",
        "Last Name",
        "Gender",
        "Age",
        "AgeGroup",

        "Email",
        "Phone Number",

        "Department",
        "Job Title",

        "Experience",
        "ExperienceLevel",
        "Projects Completed",

        "Salary",
        "SalaryLevel",

        "Training Hours",
        "TrainingCategory",

        "Attendance Rate",
        "AttendanceCategory",

        "Education",

        "Performance Score",
        "HighPerformer",

        "BonusEligibility",
    ]
]
```
